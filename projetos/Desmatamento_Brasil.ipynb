# Análise do desmatamento em alguns estados do Brasil

O presente projeto, tem por objetivo analisar um conjunto de dados sobre o desmatamento em alguns estados brasileiros. Esse dataset foi escolhido por já estar bem simplificado e, por isso, ser mais amigável a um aprendiz dem Data Science.

Foi feito o carregamento da base de dados, verificou - se a existência de valores nulos ou incoscistentes e foram plotados alguns gráficos sobre o desmatamento total em cada ano, por estado e a evolução do desmatamento em cada um deles.

```python
import pandas as pd # importando a biblioteca pandas, para carregamento e manipulações no dataset.

df_desmatamento = pd.read_csv('Data/desmatamento.csv') # carregamento do arquivo .csv com os dados.
df_desmatamento.head(10) # visualizando as dez primeiras linhas da base de dados.
```

A seguir é feita a verificação dos dados nulos ou inconscistentes.

```python
df_desmatamento.info()
```

```python
df_desmatamento.isna().sum()
```

```python
df_desmatamento.isnull().sum()
```

## Visualização dos dados

O gráfico abaixo mostra o total de desmatemento em cada ano.

```python
import matplotlib.pyplot as plt

grafico = plt.bar(df_desmatamento['referencia'], df_desmatamento['area_total_desmatamento'])
plt.title('Total de desmatamento em cada ano')
grafico;
```

O gráfico a seguir compara a quantidade total de desmatamento de cada estado.

```python
estados = list(df_desmatamento.columns[1: 10])

desmatamento_por_estado = []

for uf in estados:
  desmatamento_por_estado.append(df_desmatamento[uf].sum())

grafico2 = plt.bar(estados, desmatamento_por_estado)
plt.xticks(rotation = 45)
plt.title('Área total desmatada em cada estado')
grafico2;
```

No gráfico a seguir podemos verificar a evolução do desmatamento em cada estado.

```python
plt.figure(figsize = (12, 6))
df_desmatamento.plot(x = 'referencia', y = estados, figsize = (12, 6))

plt.title('Evolução do desmatamento em cada estado')
plt.xlabel('Ano')
plt.ylabel('Área desmatada (km²)')
plt.legend(loc='best')
plt.grid(True)
plt.show()
```
